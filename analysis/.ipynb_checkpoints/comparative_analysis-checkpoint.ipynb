```python
from pathlib import Path
import numpy as np
import pandas as pd

C = [125.0, 181.0, 39.0]

def get_omega(filepath):
    with open(filepath, 'r') as f:
        arrs = [eval(x) for x in f.read().split("+")]
    assert len(C) == len(arrs)
    q = [np.array(a) for a in arrs]
    m = [np.ceil(q[i] / C[i]).astype(int) for i in range(len(C))]
    return [q,m]

data_pso = []
pathlist = Path("/home/pc/main/rust/network-pso/soln/intermediate").glob('*')
for path in pathlist:
    data_pso.append(get_omega(path))
        
print(data_pso[0][1])
```

```python
data_cro = []
pathlist = Path("/home/pc/main/rust/network-cro/soln/intermediate").glob('*')
for path in pathlist:
    data_cro.append(get_omega(path))
        
print(data_cro[0][1])
```

```python
dist = []
dist.append(pd.read_excel(
    "/home/pc/main/rust/network-cro/resources/Data_Distance_^0_Demand_17-09-2023.xlsx",
    'Distance Matrix',
    skiprows = 2,
    nrows = 2,
    usecols = 'C:C'
))

dist.append(pd.read_excel(
    "/home/pc/main/rust/network-cro/resources/Data_Distance_^0_Demand_17-09-2023.xlsx",
    'Distance Matrix',
    skiprows = 8,
    nrows = 2,
    usecols = 'C:E'
))

dist.append(pd.read_excel(
    "/home/pc/main/rust/network-cro/resources/Data_Distance_^0_Demand_17-09-2023.xlsx",
    'Distance Matrix',
    skiprows = 13,
    nrows = 4,
    usecols = 'B:AJ'
))

print(dist[2])
```

```python
L = [2,3,1]
UL = [1.5,2.5,0.5]
s = [50,800,60]

def tdt(data,idx):
    res = 0
    for i in range(3):
        T = dist[i].to_numpy()/s[i]
        T += (L[i]+UL[i])
        res += (data[idx][1][i][:,:,0]*T).sum()
    return res

tdt(data_pso,0)
```

```python
tdt_pso = []
tdt_cro = []
demand = [529,959,1355]

for i in (0,2,1):
    tdt_pso.append(tdt(data_pso,i))
    tdt_cro.append(tdt(data_cro,i))

print(tdt_pso)
print(tdt_cro)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib_inline
import seaborn as sns

# seaborn theme defaults
matplotlib_inline.backend_inline.set_matplotlib_formats('png')
sns.set_theme()
sns.set_palette("muted")

def compare_convergence(fp_cro,fp_pso):
    with open(fp_cro, 'r') as f:
        min_f_cro = np.array([float(x) for x in f.read().split(",")])
    with open(fp_pso, 'r') as f:
        min_f_pso = np.array([float(x) for x in f.read().split(",")])
    multiplier = 300 / len(min_f_cro)
    epoch_cro = [multiplier*(i+1) for i in range(len(min_f_cro))]
    multiplier = 300 / len(min_f_pso)
    epoch_pso = [multiplier*(i+1) for i in range(len(min_f_pso))]
    sns.set_theme()
    plt.plot(epoch_cro, min_f_cro,label="CRO")
    plt.plot(epoch_pso,min_f_pso,label="PSO")
    plt.xlabel('Time (s)')
    plt.ylabel('Total Delivery Time')
    plt.legend()
    plt.show()
        
cro_fps = [
    "/home/pc/main/rust/network-cro/logs/2023-10-02T09:53:04.861753730+00:00.log",
    "/home/pc/main/rust/network-cro/logs/2023-10-02T10:02:38.637851084+00:00.log",
    "/home/pc/main/rust/network-cro/logs/2023-10-02T10:11:24.216805396+00:00.log"
]
pso_fps = [
    "/home/pc/main/rust/network-pso/logs/2023-10-08T19:20:36.940270371+00:00.log",
    "/home/pc/main/rust/network-pso/logs/2023-10-08T19:27:17.088013704+00:00.log",
    "/home/pc/main/rust/network-pso/logs/2023-10-08T19:33:34.018289767+00:00.log"
]

compare_convergence(cro_fps[0],pso_fps[0])
```

```python
def plot_tdt():
    plt.plot(demand,tdt_cro,label="CRO",marker=".")
    plt.plot(demand,tdt_pso,label="PSO",marker=".")
    plt.xlabel("Number of ventilators required")
    plt.ylabel("Total Delivery Time (h)")
    plt.legend(loc="upper left")
    plt.title("TDT Comparison of CRO and PSO algorithms")
    plt.show()

plot_tdt()
```
